```python
import pandas as pd
import string
import re

canada_edu = pd.read_csv('Canada_Education.csv')
canada_edu.isnull()
```

```python
canada_edu = canada_edu.drop('Place', axis=1)
```

```python
canada_edu.isnull()
```

```python
canada_edu = canada_edu.fillna("NaN")
```

```python
canada_edu.isnull()
```

```python
canada_edu.head(20)
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

canada_edu['Text'] = canada_edu['Text'].apply(lambda x: remove_punct(x))
canada_edu['Author'] = canada_edu['Author'].apply(lambda x: remove_punct(x))
canada_edu['Location'] = canada_edu['Location'].apply(lambda x: remove_punct(x))
#university['Date'] = university['Date'].apply(lambda x: remove_punct(x))
canada_edu.head(20)
```

```python
canada_edu['Text'] = canada_edu['Text'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
canada_edu['Location'] = canada_edu['Location'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```
