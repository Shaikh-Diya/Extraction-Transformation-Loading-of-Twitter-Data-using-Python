```python
import pandas as pd
import string
import re

halifax = pd.read_csv('Halifax.csv')
halifax.isnull() 
```

```python
halifax = halifax.drop('Place', axis=1)
```

```python
halifax.isnull()
```

```python
halifax = halifax.fillna("NaN")
```

```python
halifax.isnull()
```

```python
halifax.head(20)
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

halifax['Text'] = halifax['Text'].apply(lambda x: remove_punct(x))
halifax['Author'] = halifax['Author'].apply(lambda x: remove_punct(x))
halifax['Location'] = halifax['Location'].apply(lambda x: remove_punct(x))
#university['Date'] = university['Date'].apply(lambda x: remove_punct(x))
halifax.head(20)
```

```python
halifax['Text'] = halifax['Text'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
halifax['Location'] = halifax['Location'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```

```python
halifax.to_csv('Cleaned_Halifax_twitter.csv')
```
