```python
import pandas as pd
import string
import re

university = pd.read_csv('University.csv')
university.isnull()
```

```python
university = university.drop('Place', axis=1)
```

```python
university.isnull()
```

```python
university = university.fillna("NaN")
```

```python
university.isnull()
```

```python
university.head(20)
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

university['Text'] = university['Text'].apply(lambda x: remove_punct(x))
university['Author'] = university['Author'].apply(lambda x: remove_punct(x))
university['Location'] = university['Location'].apply(lambda x: remove_punct(x))
#university['Date'] = university['Date'].apply(lambda x: remove_punct(x))
university.head(20)
```

```python
university['Text'] = university['Text'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
university['Location'] = university['Location'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```

```python
university.to_csv('Cleaned_University_twitter.csv')
```
