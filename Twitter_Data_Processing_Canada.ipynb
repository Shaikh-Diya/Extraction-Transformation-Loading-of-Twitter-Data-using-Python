```python
import pandas as pd
import string
import re

canada = pd.read_csv('Canada.csv')
canada.isnull()
```

```python
canada = canada.drop('Place', axis=1)
```

```python
canada.isnull()
```

```python
canada =canada.fillna("NaN")
```

```python
canada.isnull()
```

```python
canada.head(20)
```

```python
string.punctuation
```

```python
def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

canada['Text'] = canada['Text'].apply(lambda x: remove_punct(x))
canada['Author'] = canada['Author'].apply(lambda x: remove_punct(x))
canada['Location'] = canada['Location'].apply(lambda x: remove_punct(x))
#university['Date'] = university['Date'].apply(lambda x: remove_punct(x))
canada.head(20)
```

```python
canada['Location'] = canada['Location'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
canada['Text'] = canada['Text'].str.replace('[^\w\s#@/:%.,_-]', '', flags=re.UNICODE)
```
