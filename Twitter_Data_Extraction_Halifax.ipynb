```python
import csv
class StreamListener(tweepy.StreamListener):
    
    def __init__(self):
        super(StreamListener, self).__init__()
        self.num_tweets=0

    def on_status(self, status):
        print(status.text)
        self.num_tweets+=1
        if(self.num_tweets<=3500):
            with open('Halifax.csv', 'a') as Canada_Education:
                writer = csv.writer(Canada_Education)
                writer.writerow([status.author.screen_name, status.created_at, status.text, status.user.location])
                
                return True
        else:
            return False
        
        
    #def on_data(self, data):
        #json_data = json.loads(data)
        #file.write(str(json_data))
         
    def on_error(self, status_code):
        if status_code == 420:
            return False

with open('Halifax.csv','w') as Canada_Education:
    writer = csv.writer(Canada_Education)
    writer.writerow(['Author', 'Date', 'Text', 'Location', 'Place'])

stream_listener = StreamListener()
stream = tweepy.Stream(auth=api.auth, listener=stream_listener)
stream.filter(track=["Halifax"],languages=["en"])
Canada_Education.close()
```
